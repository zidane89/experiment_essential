```python
import matplotlib.pyplot as plt 
import pickle 
import numpy as np 
```

```python
with open("DDPG1_400.pkl", "rb") as f: 
    data_w = pickle.load(f)

with open("DDPG2_400.pkl", "rb") as f: 
    data_norm = pickle.load(f)

with open("DDPG3_400.pkl", "rb") as f: 
    data_batchnorm = pickle.load(f)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(data_w[3]["rewards"], "k", label="power in [W]", linewidth=2)
plt.plot(data_norm[3]["rewards"], "b", label="normalized power", linewidth=2)
plt.plot(data_batchnorm[3]["rewards"], "r", label="batch normalization layer", linewidth=2)

plt.xlabel("episodes", fontsize=16)
plt.ylabel("episodic rewards", fontsize=16)
plt.legend(fontsize=14)
plt.xlim([0, 130])
plt.grid()
plt.savefig("results.png")
```
